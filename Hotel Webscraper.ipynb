```python
import requests
from bs4 import BeautifulSoup
from datetime import datetime, timedelta
import pandas as pd

start_date = datetime.strptime("2024-02-25", "%Y-%m-%d")
end_date = datetime.strptime("2024-02-26", "%Y-%m-%d")
end_date_limit = datetime.strptime("2025-01-01", "%Y-%m-%d")

headers = {
    "User-Agent": "Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/102.0.5042.108 Safari/537.36"
}

hotel_results = []

while start_date < end_date_limit:
    check_in = start_date.strftime("%Y-%m-%d")
    check_out = (start_date + timedelta(days=1)).strftime("%Y-%m-%d")

    url = f"https://www.booking.com/searchresults.html?ss=San+Antonio&ssne=San+Antonio&ssne_untouched=San+Antonio&label=gen173nr-1FCAQoggJCC3JlZ2lvbl80MTI3SDFYBGiwAogBAZgBMbgBF8gBDNgBAegBAfgBA4gCAagCA7gCy-DqrgbAAgHSAiQ4NTY4ZGYyNS0yOWY3LTQzZTQtOTI1NS1hZGMwMzNjZDdmMmbYAgXgAgE&aid=304142&lang=en-us&sb=1&src_elem=sb&src=searchresults&dest_id=20131185&dest_type=city&checkin={check_in}&checkout={check_out}&group_adults=1&no_rooms=1&group_children=0"

    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.content, 'html.parser')

    print(f"Scraping data for check-in: {check_in}, check-out: {check_out}")

    for el in soup.find_all("div", {"data-testid": "property-card"}):
        name = el.find("div", {"data-testid": "title"}).text.strip()
        link = el.find("a", {"data-testid": "title-link"})["href"]
        location = el.find("span", {"data-testid": "address"}).text.strip()
        pricing = el.find("span", {"data-testid": "price-and-discounted-price"}).text.strip()

        rating_element = el.find("div", {"data-testid": "review-score"})
        rating = rating_element.text.strip().split(" ")[0] if rating_element else "N/A"
        review_count = rating_element.text.strip().split(" ")[1] if rating_element else "N/A"

        thumbnail = el.find("img", {"data-testid": "image"})['src'] if el.find("img", {"data-testid": "image"}) else "N/A"

        hotel_results.append({
            "name": name,
            "link": link,
            "location": location,
            "pricing": pricing,
            "rating": rating,
            "review_count": review_count,
            "thumbnail": thumbnail,
            "check_in": check_in,  # Add check_in date here
        })

    # Increment start date for next iteration
    start_date += timedelta(days=1)

print("Scraping completed.")

# Convert the list of dictionaries to a pandas DataFrame
df = pd.DataFrame(hotel_results)

# Save the DataFrame to a CSV file including the check_in column
df.to_csv('hotels_final_DEADPOOL.csv', index=False, encoding='utf-8-sig')
```

```python
import pandas as pd

# Step 1: Load the dataset
file_path = 'hotels_historical_price.csv'  # Update this to your file path
data = pd.read_csv(file_path)

# Step 2: Pre-process the 'pricing' column
data['pricing'] = data['pricing'].str.replace('$', '').astype(float)

# Step 3: Calculate the average price by date
avg_price_by_date = data.groupby('check_in')['pricing'].mean().reset_index(name='avg_price')

# Step 4: Store in a new DataFrame (already done in step 3)

# Step 5: Save the new DataFrame to a CSV file
output_file_path = 'avg_price_by_date.csv'  # Update this to your desired output path
avg_price_by_date.to_csv(output_file_path, index=False)

print(f"File saved to {output_file_path}")
```
